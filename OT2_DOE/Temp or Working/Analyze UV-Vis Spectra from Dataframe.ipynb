```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_data = pd.read_csv(r"C:\Users\Edwin\Desktop\05_07_21_5CB_Abs.txt")
full_data
```

```python
wavelength_columns = [col for col in full_data.columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]
spectras = full_data[wavelength_columns]
```

```python
I0_over_IT = 10**spectras
IT_over_I0 = 1/I0_over_IT
turb = np.log(IT_over_I0)/-0.8
turb
```

```python
oelic_acid = full_data['OleicAc concentration molarity']
oleyamine = full_data['Oleylamine concentration molarity']
sonication_time = full_data['Total Sonication Time Min']

labels = ['Oelic Acid = ' str(oa) for oa, ol, t in zip(oelic_acid)]
```

```python
# labels = full_data['OleicAc concentration molarity']
for j, (i, row) in enumerate(turb.iterrows()):
    row = list(row)
    full_data_row = full_data.iloc[i]
    plt.plot(wavelengths, row)
# plt.xlim(350,450)
# plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```

```python
selected_data = full_data[full_data['Total Sonication Time Min']==0]
label_1= selected_data['Oleylamine concentration molarity']
label_2 = selected_data['OleicAc concentration molarity']
labels = ['Oleylamine Molairty =  '+ str(l1) + '\n Oleic Acid Molarity = ' + str(l2) for l1, l2 in zip(label_1, label_2)]
labels
```

```python
# print(len(labels))
selected_spectras = selected_data[wavelength_columns]
selected_spectras
for j, (i, row)in enumerate(selected_spectras.iterrows()):
    row = list(row)
    plt.plot(wavelengths, row, label = labels[j])
    
# plt.xlim(370,700)
# plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```
