```python
def one_to_one_transfer(protocol, volume_1, volume_2, source_wells, destination_wells_1, destination_wells_2, loaded_dict): # figure out how kawrgs work so you can feed it.
    small_pipette = loaded_dict['Small Pipette']
    large_pipette = loaded_dict['Large Pipette']
    
    if small_pipette.min_volume <= volume_1 <= large_pipette.max_volume:
        pipette = small_pipette
    elif volume_1 >= large_pipette.min:
        pipette = large_pipette
    else:
        pass
    
    
    for i in range(len(source_wells)):
        pipette.pick_up_tip()
        pipette.transfer(volume_1, source_wells[i], destination_wells_1[i], new_tip='never', mix_before = (1, pipette.max_volume))
        pipette.transfer(volume_2, source_wells[i], destination_wells_2[i], new_tip='never')
        pipette.drop_tip()
    
    for line in protocol.commands(): 
        print(line) 
```

```python
one_to_one_transfer(protocol, 300, 30, sample_wells[:81], undiluted_plate_wells[:81], diluted_plate_wells[:81], loaded_dict)
```
