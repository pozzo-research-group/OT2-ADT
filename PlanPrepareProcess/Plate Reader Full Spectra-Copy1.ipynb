```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_data = pd.read_csv(r"03_16_21_Synthesis_Plus_Spectra")
# full_data = full_data.round(4) # issue since there is some samples with less than 4 sigfigs bt the nthe unique ethanols
wavelength_columns = [col for col in full_data.columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]

# utilize the recursive nature of the dataframes to blank rather than rewriting
```

```python
samples = full_data[0:58]
samples.loc[:,'ethanol wtf'] = samples['ethanol wtf'].round(3)
samples_spectra = samples[wavelength_columns]

blanks = full_data[(full_data['pfh wtf']==0) & (full_data['dppc wtf']==0) & (full_data['dspepeg200 wtf']==0)]
blanks_spectra = blanks[wavelength_columns]

unique_ethanol_samples = samples['ethanol wtf'].unique()
unique_ethanol_blanks = blanks['ethanol wtf'].unique()
# check if both are same, hmm need to have something that iterates through a row and checks what it it needs to subtract.
samples
```

```python
blanked_sample_spec = []
for sample_conc, blank_conc in zip(unique_ethanol_samples, unique_ethanol_blanks): #essentially to check enough blank concs exist to cover the samples
    if sample_conc == blank_conc: 
        unblanked_full = samples[samples['ethanol wtf']==blank_conc]
        unblanked_spectra = samples_spectra.loc[unblanked_full.index]
        
        blank_full = blanks[blanks['ethanol wtf']==blank_conc]
        blank_spectra = blanks_spectra.loc[blank_full.index]
                  
        blanked_spectra = unblanked_spectra - blank_spectra.values
        blanked_sample_spec.append(blanked_spectra)
    else: 
        print('check')
        
pd.concat(blanked_sample_spec)
```

```python
samples[wavelength_columns] = pd.concat(blanked_sample_spec) # this is so jank
```

```python
# blanked_samples_spectra = samples[wavelength_columns]
samples.to_csv(r"03_16_21 Synthesis Info Plus Blanked Spectras", index=False)
samples
```
