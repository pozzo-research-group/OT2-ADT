```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from Process import PlateReader
```

```python
path_1 = r"C:\Users\Edwin\Downloads\03_30_21_7pt5min.xlsx"
data_1 = pd.read_excel(path_1, sheet_name='Sheet1').T
spectra_1 = data_1.iloc[1:]
wavelengths = data_1.loc['Wavelength'].tolist()

path_2 = r"C:\Users\Edwin\Downloads\03_30_21_15min.xlsx"
spectra_2 = pd.read_excel(path_2, sheet_name='Sheet1').T.iloc[1:]
blank = spectra_2.iloc[-1:]

path_3 = r"C:\Users\Edwin\Downloads\03_30_21_22pt5min.xlsx"
spectra_3 = pd.read_excel(path_3, sheet_name='Sheet1').T.iloc[1:]
spectra_3 = PlateReader.detect_ovflw(spectra_3, holder=15)

path_4 = r"C:\Users\Edwin\Downloads\03_30_21_30min.xlsx"
spectra_4 = pd.read_excel(path_4, sheet_name='Sheet1').T.iloc[1:]
spectra_4 = PlateReader.detect_ovflw(spectra_4, holder=15)
```

```python
# nsupported operand type(s) for -: 'str' and 'float' means probably you have a float somewhere
```

```python
spectra1_blanked = spectra_1 - blank.loc[blank.index.repeat(len(spectra_1))].values
spectra2_blanked = spectra_2 - blank.loc[blank.index.repeat(len(spectra_2))].values
spectra3_blanked = spectra_3 - blank.loc[blank.index.repeat(len(spectra_3))].values
spectra4_blanked = spectra_4 - blank.loc[blank.index.repeat(len(spectra_4))].values
```

```python
for i, row in spectra1_blanked.iterrows():
    row = list(row)
    plt.plot(wavelengths, row, label='7.5 min')
    
plt.legend()    
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
plt.xlim(200,500)
```

```python
for i, row in spectra1_blanked.iterrows():
    row = list(row)
    plt.plot(wavelengths, row, label = '7.5 min')
    
for i, row in spectra2_blanked[:-1].iterrows():
    row = list(row)
    plt.plot(wavelengths, row, label = '15 min')
    
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
plt.xlim(350,450)
plt.ylim(0,0.4)
```

```python
fig, ax = plt.subplots()
ax.plot(wavelengths, spectra1_blanked.iloc[0].tolist(), label ='7.5 minutes')
ax.plot(wavelengths, spectra2_blanked.iloc[0].tolist(), label ='15 minutes')
ax.plot(wavelengths, spectra3_blanked.iloc[0].tolist(), label ='22.5 minutes')
ax.plot(wavelengths, spectra4_blanked.iloc[0].tolist(), label ='30 minutes')
ax.legend()
ax.set_xlim(375,475)
ax.set_ylim(0,1)
ax.set_xlabel('Wavelength nm')
ax.set_ylabel('AU')
```
