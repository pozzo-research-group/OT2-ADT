```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_data = pd.read_csv(r"04_09_21 Cadmium Oleate HT Blanked")
```

```python
wavelength_columns = [col for col in full_data.columns if 'nm' in col]
wavelengths = [float(wavelength.replace("nm","")) for wavelength in wavelength_columns]
spectras = full_data[wavelength_columns]
```

```python
for i, row in spectras.iterrows():
    row = list(row)
    plt.plot(wavelengths, row)
plt.xlim(350,450)
plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```

```python
selected_data = full_data[full_data['ZnOl 84mM volpct'] == 0.5]
labels = selected_data['Time Sonicated']
selected_spectras = selected_data[wavelength_columns]

for i, row in selected_spectras.iterrows():
    row = list(row)
    plt.plot(wavelengths, row, label = labels[i])
    
plt.xlim(350,450)
plt.ylim(-0.05, 1)
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('AU')
```
